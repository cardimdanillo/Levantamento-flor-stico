```python
import pandas as pd
import numpy as np
```

```python
from flora_volks import dados_com_gen

#display(dados_com_gen)
```

```python
dados_com_gen['Número'] = dados_com_gen['Número'].astype('str')
dados_com_gen.info()
```

```python
tomboUB = pd.read_excel("tomboUB.xlsx")
tomboUB = tomboUB.rename({'number':'Número'}, axis=1)

indice_col = tomboUB.columns.get_loc('collector')
tomboUB = tomboUB.iloc[:, :indice_col + 1]

tomboUB['Número'] = tomboUB['Número'].astype('str')

print(tomboUB.info())
display(tomboUB)
```

```python
dados_com_gen = dados_com_gen.merge(tomboUB, on = 'Número', how='left')
```

```python
#Agrupas por espécies e agregar número de coletor e codbarras
columns = ['Família','Gênero','sp','autor','Hábito','Origem','Substrato','Endemismo','collector']

dados_agr_rel = dados_com_gen.groupby(columns).agg({'Número': ', '.join, 'codbarras': ', '.join}).reset_index()
#dados_agr_rel = dados_com_gen.groupby(columns).agg({'Número': ', '.join}).reset_index()

display(dados_agr_rel)
```

```python
dados_agr_rel = dados_agr_rel.sort_values(by=['Família','Gênero','sp'])
dados_agr_rel.loc[dados_agr_rel.Família == 'Piperaceae']
```

```python
dados_agr_rel['Espécie'] = dados_agr_rel['Gênero'].str.cat(dados_agr_rel['sp'], sep=" ").str.cat(dados_agr_rel['autor'],sep=" ")
drop_columns = ['sp','Gênero','autor']
dados_agr_rel = dados_agr_rel.drop(labels=drop_columns, axis=1)
```

```python
new_column_order = ['Número', 'Família', 'Espécie','Hábito','Origem','Substrato','Endemismo','codbarras','collector']
dados_agr_rel = dados_agr_rel.reindex(columns=new_column_order)
print(dados_agr_rel)

display(dados_agr_rel)
```

```python
dados_agr_rel['collector'].unique()
```

```python
dados_agr_rel = dados_agr_rel.replace({'Ferreira, IP':'Ferreira, IP et al.','Carvalho-Silva, M':'Carvalho-Silva, M et al.'})
```

```python
nome_arquivo_excel = 'FloraVolks_apendiceA.xlsx'

dados_agr_rel.to_excel(nome_arquivo_excel, index=False)
```
